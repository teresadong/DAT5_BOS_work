```python
%matplotlib inline
# importing division from the 'future' release of Python (i.e. Python 3)
from __future__ import division

import pandas
import sqlite3

# connect to the baseball database. Notice I am passing the full path
# to the SQLite file.
database = r'C:\Users\Teresa\Documents\GA Data Science\lahman2013.sqlite'
conn = sqlite3.connect(database)
# creating an object contraining a string that has the SQL query. 
sql = """SELECT hfi.playerID,hfi.inducted, batting.*, pitching.*, fielding.* FROM 
        (SELECT playerID, CASE WHEN average_inducted = 0 then 0 else 1 end as inducted from 
            (
            select playerID, avg(case when inducted = 'Y' then 1 else 0 end ) as average_inducted from  HallOfFame hf
            where yearid < 2000
            group by playerID
            )
        ) hfi 
    LEFT OUTER JOIN
        (
        select playerID,  
        sum(AB) as batter_at_bats, 
        sum(H) as batter_hits, 
        sum(R) as batter_runs, 
        sum(HR) as batter_home_runs, 
        sum(SB) as batter_stolen_bases, 
        sum(RBI) as batter_RBI, 
        sum(CS) as batter_caught_stealing, 
        sum(SO) as batter_strikeouts, 
        sum(IBB) as batter_intentional_walks
        from Batting
        group by playerID
        ) batting on batting.playerID = hfi.playerID
    LEFT OUTER JOIN
        (
         select playerID, 
         sum(G) as games_pitched, 
         sum(SO) as pitcher_shutouts, 
         sum(sv) as pitcher_saves, 
         sum(IPouts) as outs_pitched, 
         sum(H) as pitching_hits, 
         sum(er) as pitching_earned_runs, 
         avg(ERA) as average_ERA, 
         sum(WP) as wild_pitches, 
         sum(HBP) as hit_by_pitch, 
         sum(GF) as pitcher_games_finished,
         sum(R) as pitcher_runs_allowed
        from Pitching
        group by playerID
        ) pitching on pitching.playerID = hfi.playerID 
    LEFT OUTER JOIN
        (
        select playerID, 
        sum(G) as games_fielded, 
        sum(InnOuts) as time_in_field_with_outs, 
        sum(PO) as field_putouts, 
        sum(E) as field_errors, 
        sum(DP) as field_double_plays
        from Fielding
        group by playerID
        ) fielding on fielding.playerID = hfi.playerID""";
# passing the connection and the SQL string to pandas.read_sql.
df = pandas.read_sql(sql, conn)
# closing the connection.
conn.close()



# dropping ALL NaNs in the dataset.
df.dropna(inplace = True)

print df.head()
```

```python
#removing extra playerID columns
df.drop('playerID',  1, inplace = True)

#rename columns
df.rename(columns={'hfi.playerID': 'playerID', 'hfi.inducted': 'inducted'}, inplace=True)
```

```python
#It appears as variables are added the model error rate decreases, with the complete model having the lowest error rate at 1%
#Time to Cross Validate the models
#Use Scikit Learns logistic regression

import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from patsy import dmatrices
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
```

```python
# create dataframes with an intercept column and dummy variables for
# occupation and occupation_husb
print 'First Model  - Single Variable Predicting Inducted using outs_pitched'
y, X = dmatrices('inducted ~ outs_pitched',df, return_type="dataframe")
print X.columns

# flatten y into a 1-D array
y = np.ravel(y)

# instantiate a logistic regression model, and fit with X and y
model = LogisticRegression()
model = model.fit(X, y)

# check the accuracy on the training set
print model.score(X, y)
#Accuracy is 89%

# what percentage were inducted
print y.mean()
#It appears that 14% of players were inducted

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=10)
print scores
print scores.mean()

#With cross validation we see that the model is performing at 90% accuracy
```
