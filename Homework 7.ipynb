```python
%matplotlib inline
# importing division from the 'future' release of Python (i.e. Python 3)
from __future__ import division

import pandas
import sqlite3

# connect to the baseball database. Notice I am passing the full path
# to the SQLite file.
database = r'C:\Users\Teresa\Documents\GA Data Science\lahman2013.sqlite'
conn = sqlite3.connect(database)
# creating an object contraining a string that has the SQL query. 
sql = """SELECT hfi.playerID,hfi.inducted, batting.*, pitching.*, fielding.* FROM 
        (SELECT playerID, CASE WHEN average_inducted = 0 then 0 else 1 end as inducted from 
            (
            select playerID, avg(case when inducted = 'Y' then 1 else 0 end ) as average_inducted from  HallOfFame hf
            where yearid < 2000
            group by playerID
            )
        ) hfi 
    LEFT OUTER JOIN
        (
        select playerID,  
        sum(AB) as batter_at_bats, 
        sum(H) as batter_hits, 
        sum(R) as batter_runs, 
        sum(HR) as batter_home_runs, 
        sum(SB) as batter_stolen_bases, 
        sum(RBI) as batter_RBI, 
        sum(CS) as batter_caught_stealing, 
        sum(SO) as batter_strikeouts, 
        sum(IBB) as batter_intentional_walks
        from Batting
        group by playerID
        ) batting on batting.playerID = hfi.playerID
    LEFT OUTER JOIN
        (
         select playerID, 
         sum(G) as games_pitched, 
         sum(SO) as pitcher_shutouts, 
         sum(sv) as pitcher_saves, 
         sum(IPouts) as outs_pitched, 
         sum(H) as pitching_hits, 
         sum(er) as pitching_earned_runs, 
         avg(ERA) as average_ERA, 
         sum(WP) as wild_pitches, 
         sum(HBP) as hit_by_pitch, 
         sum(GF) as pitcher_games_finished,
         sum(R) as pitcher_runs_allowed
        from Pitching
        group by playerID
        ) pitching on pitching.playerID = hfi.playerID 
    LEFT OUTER JOIN
        (
        select playerID, 
        sum(G) as games_fielded, 
        sum(InnOuts) as time_in_field_with_outs, 
        sum(PO) as field_putouts, 
        sum(E) as field_errors, 
        sum(DP) as field_double_plays
        from Fielding
        group by playerID
        ) fielding on fielding.playerID = hfi.playerID""";
# passing the connection and the SQL string to pandas.read_sql.
df = pandas.read_sql(sql, conn)
# closing the connection.
conn.close()



# dropping ALL NaNs in the dataset.
df.dropna(inplace = True)

print df.head()
```

```python
#removing extra playerID columns
df.drop('playerID',  1, inplace = True)

#rename columns
df.rename(columns={'hfi.playerID': 'playerID', 'hfi.inducted': 'inducted'}, inplace=True)


#get columns
col_names = list(df.columns.values)
col_names.remove('playerID')
col_names.remove('inducted')
print col_names

```

```python
import numpy as np
import pandas
from statsmodels.formula.api import logit
from statsmodels.nonparametric import kde
import matplotlib.pyplot as plt
from patsy import dmatrix, dmatrices
```

```python
#Lets try regression with just one predictor variable
model1 = logit('inducted ~ batter_hits', data = df).fit() # model1 is our fitted model.
print model1.summary()
```
